```python
import pandas as pd
import glob
import os
import time
start=time.time()
TIMESTAMP_LST=[]
SYMBOL_LST=[]
OPEN_LST=[]
HIGH_LST=[]
LOW_LST=[]
CLOSE_LST=[]
TOTTRDQTY_LST=[]
TOTTRDVAL_LST=[]
TOTALTRADS_LIST=[]
DELIVERYQNTY_LIST=[]
DELPER_LIST=[]
currentdir=os.getcwd()
#--------------DAT TO CSV-----------------------------------------
os.chdir(r'C:\Users\gandhig\Desktop\cash_data\DELIVERY') 
txt_file_list=glob.glob('*.DAT')
total_file=len(txt_file_list)
for i in range(0,total_file):
    os.rename(txt_file_list[i],txt_file_list[i].replace('.DAT','.csv'))
csv_file_list=glob.glob('*.csv')
total_csv_file=len(csv_file_list)
#----------------CSV DATA FILTERING----------------------------------
for j in range(0,total_csv_file):
    df=pd.read_csv(csv_file_list[j],skiprows=4)
    df.to_csv('temp.csv',index=False)
    newdf=pd.read_csv('temp.csv')
    newdf.columns=['Record Type','Sr No','Name of Security','SecurityType','Deliverable Quantity(gross across client level)','% of Deliverable Quantity to Traded Quantity','Deliverable%']
    newdf.to_csv(csv_file_list[j],index=False)
os.remove('temp.csv')

#-------------zip file list----------------------------------------
os.chdir(r'C:\Users\gandhig\Desktop\cash_data\ZIP') 
zip_file_list=glob.glob('*.zip')
total_zip_file=len(zip_file_list)
dff=pd.read_csv(zip_file_list[0])
symbol_list=dff['SYMBOL'][dff.SERIES=='EQ'].tolist()
total_symbol=len(symbol_list)

#----------delivery file list------------------------------------
os.chdir(r'C:\Users\gandhig\Desktop\cash_data\DELIVERY') 
csv_file_list=glob.glob('*.csv')
total_csv_file=len(csv_file_list)

def data_process(symbol):
    for i  in range(0,total_zip_file):
        os.chdir(r'C:\Users\gandhig\Desktop\cash_data\ZIP') 
        df=pd.read_csv(zip_file_list[i])
        for j in range(0,1): 
            time=df['TIMESTAMP'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            symboll=df['SYMBOL'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            OPEN=df['OPEN'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            high=df['HIGH'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            low=df['LOW'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            close=df['CLOSE'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            tottrdqty=df['TOTTRDQTY'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            tottrdval=df['TOTTRDVAL'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            totaltrades=df['TOTALTRADES'][(df.SYMBOL==symbol) & (df.SERIES=='EQ')].to_string(index=False)
            TIMESTAMP_LST.append(time)
            SYMBOL_LST.append(symboll)
            OPEN_LST.append(OPEN)
            HIGH_LST.append(high)
            LOW_LST.append(low)
            CLOSE_LST.append(close)
            TOTTRDQTY_LST.append(tottrdqty)
            TOTTRDVAL_LST.append(tottrdval)
            TOTALTRADS_LIST.append(totaltrades)  
     
    os.chdir(r'C:\Users\gandhig\Desktop\cash_data\DELIVERY')      
    for k in range(0,total_csv_file):
        df1=pd.read_csv(csv_file_list[k])
        df1.columns=['RecordType','SrNo','NameofSecurity','SecurityType','DeliverableQuantity(gross across client level)','%ofDeliverableQuantitytoTradedQuantity','Deliverable%']
        delqnty=df1['%ofDeliverableQuantitytoTradedQuantity'][(df1.NameofSecurity==symbol) & (df1.SecurityType=='EQ')].to_string(index=False)
        delper=df1['Deliverable%'][(df1.NameofSecurity==symbol) & (df1.SecurityType=='EQ')].to_string(index=False)
        DELIVERYQNTY_LIST.append(delqnty)
        DELPER_LIST.append(delper)
    data={'TIME':TIMESTAMP_LST,'SYMBOL':SYMBOL_LST,'OPEN':OPEN_LST,'HIGH':HIGH_LST,'LOW':LOW_LST,'CLOSE':CLOSE_LST,'TOTALTRADEQNTY':TOTTRDQTY_LST,
          'TOTALTRADEVAL':TOTTRDVAL_LST,'TOTALTRADES':TOTALTRADS_LIST,'DELIVERYQNTY':DELIVERYQNTY_LIST,'DELIVERY%':DELPER_LIST}
    df2=pd.DataFrame(data)
    df2.to_csv('../'+str(symbol)+str('.csv'),mode='a',index=False,header=True)
    TIMESTAMP_LST.clear()
    SYMBOL_LST.clear()
    OPEN_LST.clear()
    HIGH_LST.clear()
    LOW_LST.clear()
    CLOSE_LST.clear()
    TOTTRDQTY_LST.clear()
    TOTTRDVAL_LST.clear()
    TOTALTRADS_LIST.clear()
    DELIVERYQNTY_LIST.clear()
    DELPER_LIST.clear()     
for z in range(0,total_symbol):
    data_process(symbol_list[z])
    print(symbol_list[z])
stop=time.time()
print(stop-start)
         
  
            
            
            
            
            
```
